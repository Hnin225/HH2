```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

tokens = tokenizer.tokenize(sequence)
ids = tokenizer.convert_tokens_to_ids(tokens)

input_ids = torch.tensor([ids])  # 배치 차원을 추가
output = model(input_ids)
```

```python
sequences = ["Hello, world!", "Hugging Face is amazing!"]
batch = tokenizer(sequences, padding=True, truncation=True, return_tensors="pt")
output = model(**batch)
```

## Batch의 주요 구성 요소

### **1. input_ids**
- 각 텍스트 시퀀스를 토큰화하여 모델의 어휘(vocabulary) 사전에 해당하는 ID로 변환한 결과입니다.

#### 예시:
- `"Hello, world!"` → `[101, 7592, 1010, 2088, 999, 102, 0]`
  - `[101]`: 문장의 시작 토큰.
  - `[7592, 1010, 2088]`: `"Hello,"`, `"world!"` 단어의 토큰 ID.
  - `[102]`: 문장의 끝 토큰.
  - `[0]`: 패딩 토큰.
- `"Hugging Face is amazing!"` → `[101, 17662, 2227, 2003, 6429, 999, 102]`
  - 패딩 없이 길이가 같음.

---

### **2. attention_mask**
- **`1`**: 실제 데이터가 있는 부분.
- **`0`**: 패딩된 부분을 나타냅니다.

#### 예시:
- `"Hello, world!"` → `[1, 1, 1, 1, 1, 1, 0]` (마지막 `0`은 패딩을 의미).
- `"Hugging Face is amazing!"` → `[1, 1, 1, 1, 1, 1, 1]` (패딩 없음).

---

### **정리**
- **`input_ids`**: 모델이 이해할 수 있는 숫자 형태로 변환된 토큰 ID.
- **`attention_mask`**: 패딩 여부를 나타내는 마스크로, 모델이 패딩된 부분을 무시하도록 돕습니다.
